# Paradigma de la Programacion Orientada a Objetos (POO)

La poo es un paradigma que organiza el software en objetos, los cuales son instancias de clases.Este enfoque permite modelar entidades del mundo real mediante abstracciones que encapsulan `datos` y `comportamientos`.Python soporta plenamente la POO desde su nucleo, ofreciendo caracteristicas como :

- Herencia
- Encapsulamiento
- Polimorfismo
- Abstraccion

# ¿Que es una clase en Python?

En python, una **clase** es un objeto de primer orden que define una **plantilla** (blueprint) para crear instancias conocidas como **objetos**.

Desde una perspectiva conceptual, una clase encapsula datos (atributos)  y comportamientos (metodos) que caracterizan a una categoria de entidades del mundo real o abstracto.

Desde una perspectiva tecnica, una clase es una estructura ejecutable que, al ser incocada, produce una instancia (es decir, una clase en python es una fabrica de objetos).

# Proposito y uso de clases

## Modelado de entidades del dominio

Uno de los logros mas poderosos de la POO es su capacidad de traducir problemas del mundo real en estructuras computacionales coherentes y mantenibles. Este proceso es conocido como modelado de dominio.




### Abstraccion como mecanismo central

El modelado comienza con la abstraccion, el acto de identificar que atributos y comportamientos son relevantes para el problema en cuestion e ignorar los detalles irrelevantes.

Por ejemplo : al modelar a un  `Usuario` de un sistema de e-commerce, nos interesan su nombre, su correo, historial de compras y su metodo de autenticacion, pero no nos interesa : su color favorito, no nos interesa su formacion profesional, no nos interesa el colegio donde acabo la secundaria (a menos que el dominio asi lo requiera)

````python
# La clase es la unidad sintactica que materializa esta abstraccion
class Usuario:
  """
  Al no colocar clase padre alguna en la deficion de la clase Usuario, se interpreta
  que la clase Usuario es clase hija directa de la clase object (clase padre de todo en python)
  """

  def __init__(self, nombre, email, historial_compras, metodo_autenticacion, fecha_registro):
    """
    Documentacion la naturaleza de los elementos de cada instancia de la clase
    """

    # atributos de instancia
    self.nombre = nombre
    self.email = email
    self.historial_compras = historial_compras
    self.metodo_autenticacion = metodo_autenticacion
    self.fecha_registro = "resultado de una consulta a una BD del sistema"
    self.carrito = []
    self.activo = True

  # Luego de 3 semanas de debate, de escribir documentacion, decidir que patron de diseño utilizar
  def agregar_al_carrito(self, producto):
    """
    Documentacion como un docstring
    """
    if self.activo:
      self.carrito.append(producto)
    else:
      raise ValueError("Usuario INACTIVO")


````

### Representacion de relaciones entre entidades del dominio

El modelado no termina en clases aisladas (Usuario), el verdadero poder de las clases emerge cuando se representan relaciones entre entidades

#### Asociaciones : Dos clases estan relacionadas porque una necesita interactuar con la otra

````python
class Producto:
  def __init__(self, nombre, precio):
    self.nombre = nombre
    self.precio = precio

class Pedido:
  def __init__(self, usuario):
    self.usuario = usuario # Asociacion con una instancia cualquiera de la clase Usuario
    self.productos = [] # Lista de instancias de la clase Producto
````

#### Composicion y Agregacion

- Composicion : Una entidad "posee" otra forma  exclusiva (si el todo desaparece las partes tambien). Ejemplo : Un `Documento` esta compuesta de `Parrafos`

- Agregacion : Una relacion mas debill de `uso` o `contencion`. Ejemplo : Una clase `Departamento` (representa a un departamento en particular dentro de la estructura de una empresa) agrega instancias de una clase `Empleados` (representa a una persona dentro de la empresa), pero justamente estas instancias de la clase `Empleados` pueden existir sin instancias de la clase `Departamento`

#### Herencia y Especializacion : Cuando existen variantes de una entidad, se usan jerarquias



````python
from abc  import ABC, abstractmethod

class Pago(ABC):
  @abstractmethod
  def procesar(self):
    """
    docstring de documentacion
    """

    # logica de negocio

# Necesito definir formas de pago : PagoTarjeta PagoYape PagoPayPal
class PagoTarjeta(Pago):
  def procesar(self):
    return "procesando con tarjeta N° XXXXXX"
class PagoYape(Pago):
  def procesar(self):
    return "Procesando el pago con la red yape"
class PagoPayPal(Pago):
  def procesar(self):
    return "Procesando con cuenta PayPal N° : YYYYYY"
````

## Ventajas del modelado orientado a objetos del dominio

- Legibilidad y expresividad : El codigo se asemeje al lenguaje del dominio (Diseño Guiado por el dominio - DDD). Tanto un desarrollador y un experto del negocio puedan leer ,interpretar y entender su significado sin ambiguedades.

- Mantenibilidad : Los cambios en los requisitos de dominio (por ejemplo : "ahora los usuarios pueden tener multiples direcciones de envio") se implementan modificando solo las clases afectadas (`Usuario`) sin tocar la logica de otras entidades del dominio (como por ejemplo, la clase Facturacion, Inventario)

- Reutilizacion : Una clase bien modelada (`Producto`, `Cliente`, `Transaccion`) puede reutilizarse en multiples contextos: frontend, backend, microservicios, pruebas, etc

- Validacion y Coherencia : El estado de una entidad puede validarse en el momento de su creacion o en alguna modificacion

```python
class Email:
  def __init__(self, valor):
    if ("@" not in valor)  and ("." not in valor):
      raise ValueError("Email Invalido")
    self.valor = valor
```

- Soporte para pruebas unitarias : Cada entidad debe de probarse de forma aislada

# Desafios y buenas practicas en el modelado de dominio

## Evitar "anemicos" modelos de dominio

Un antipatron comun es el **modelo anemico**, donde las clases solo contienen datos (atributos) y toda la logica del negtocio reside en servicios externos. Esto viola la esencia de la POO.



````python
# Mala practica

class Usuario:
  def __init__(self, nombre,activo):
    self.nombre = nombre
    self.activo = activo

# Toda la logica esta fuera de la clase
def agregar_al_carrito(usuario, producto):
  if not usuario.activo:
    raise ValueError("...")
  usuario.carrito.append(producto)
````

````python
# Buena Practica
class Usuario:
  # luego del constructor
  def agregar_al_carrito(self, producto):
    if not self.activo:
      raise ValueError("...")
    self.carrito.append(producto)
````

## Preferir entidades ricas (en comportamiento)

Las clases del dominio deben  ser ricas en comportamiento, no solo en datos. Deben encapsular reglas de negocio, validaciones  e invariantes.

## Delimitar contextos con coherencia

En sistemas grandes (donde de manera muy natural tenemos muchisimas clases) es recomendable seguir los principios del **Diseño Guiado por el Dominio (Domain-Driven Design)** , identificando contextos delimitados (bounded context) donde una clase tenga un significado claro y consistente. Por ejemplo : Una clase `Producto`, cual en el contexto de un inventario (el cual puede tener todo un ecosistema de entidades) puede tener atributos distintos que en el contexto de marketing
