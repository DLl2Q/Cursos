## Atributos de instancia

Los atributos de instancia son datos que pertenecen a una instancia especifica de una clase. Cada objeto tiene su propia copia independiente
de estos atributos.

Estos atributos (generalmente) se definen en el metodo `__init__` usando `self`

```python
class EstudiantePit:
  def __init__(self, dni):
    # Atributos de instancia
    self.dni = dni
    self.notas = []

```

Caracteristicas clave:
- Unicos por objeto
- Se almacenan en el diccionario interno de la instancia (`__dict__`)
- Se crean en tiempo de instanciacion

````python

class EstudiantePit:
  def __init__(self, dni):
    # Atributos de instancia
    self.dni = dni
    self.notas = []
````

## Atributos de clase

Los atributos de clase son datos compartidos por todas las instancias de una clase. Se definen directamente en el cuerpo de la clase (fuera de cualquier metodo)

# Encapsulamiento en Python

Su proposito es : Ocultar la implementacion interna de un objeto y solo exponer una interfaz controlada para interactuar con él

## Niveles de visibilidad

Estos niveles se dan por convenciones de nomenclatura

`atributo` : Publico (Intencion : parte de la interfaz publica)

`_atributo` : Protegido (Intencion : Uso interno : no es parte de la api publica)

`__atributo` : Privado (Estrictamente interno, Intencion : Sea interno incluse frente a subclases)






## Atributos publicos

- Son accesibles desde cualquier lugar
- Forman parte de la interfaz publica de la clase

````python
class Coche:
  def __init__(self, marca):
    self.marca = marca # publico

coche4 = Coche("Volvo")
print(coche4.marca)
# Este acceso es permitido y esperado
````

## Atributos protegidos

- El guion bajo inicial (*_atributo*) es una señal para otros programadores
- No hay un bloqueo tecnico : Python no impide acceder ni modificar `_atributo`

````python
# Implementemos una clase CuentaBancaria cuyo atributo que almacene la informacion
# del saldo de la cuenta sea protegido

class CuentaBancaria:
  def __init__(self, saldo_inicial):
    self._saldo = saldo_inicial # atributo protegido

  def depositar(self, monto):
    if monto>0:
      self._saldo = self._saldo + monto
````

````python
# Creacion de una cuenta bancaria usando la clase CuentaBancaria

Miriam = CuentaBancaria(100)
````

````python
# en el codigo fuente y en la documentacion queda bastante claro que
# el atributo _saldo es protegido
#
# tecnicamente puedo modificar ese atributo _saldo pero no es recomendado
Miriam._saldo = 45
# Rompe la logica del negocio
````

````python
print(Miriam._saldo)
````

Cual es la idea de usar la convencion para definir un atributo protegido:

- Para indicar que el atributo (`_saldo`) no debe usarse directamente

- Ideal para atributos internos que podrian cambiar sin aviso

- Herramientas como linters (pylint, flake8) y entornos de desarrollo integrado porque advierten que usar `_atributo` no es una buena practica desde fuera

## Atributos privados

Cuando un atributo comienza con dos guiones bajos (`__`) y no termina con dos guiones, python aplica el mecanismo llamado `name mangling` (ofuscacion de nombre)

> El interprete renombrara internamente el atributo a `_NombreClase__atributo`

Esto no lo hace inaccesible, pero dificulta el accedo accidental , especialmente en herencia.


````python
# Implementemos una clase CuentaBancaria cuyo atributo que represente
# al numero de cuenta sea privado

class CuentaBancaria():
  def __init__(self, numero):
    self.__numero = numero # atributo privado

Gerardo = CuentaBancaria("12345")
````

````python
print(Gerardo.__numero)
````

````python
# Acceso a un atributo privado :
print(Gerardo._CuentaBancaria__numero)
````

````python
#
dir(Gerardo)
````

### ¿por que existe un atributo privado?

- Evitar colisiones de herencia

- Da una señal mas fuerte de que el atributo es interno y no debe tocarse.

**Limitaciones**
- No es seguridad real
- No funciona si el nombre termina con `__`


# Decoradores en Python

En POO en Python, a menudo queremos controlar el acceso a los atributos internos de un objeto. Python ofrece una solucion elegande y pythonica : Las `propiedades`

Una propiedad permite que un metodo se comporte como un atributo manteniendo al mismo tiempo la posibilidad de incluir logica de acceso, valicacion o calculo.

Esto se logra mediante el decorador incorporado : `@property` y sus compañeros `@atributo.setter` y `@atributo.deleter`





````python
# Sintaxis basica

class Clase:
  def __init__(self, valor):
    # atributo protegido
    self._atributo = valor

  @property
  def atributo(self):
    """
    Getter : se llama al acceder a obj.atributo
    """
    return self._atributo

  @atributo.setter
  def atributo(self, valor):
    """
    setter : se llama al hacer obj.atributo = valor
    """
    if self._validar(valor):
      self._atributo = valor
    else:
      raise ValueError("Valor invalido")

  @atributo.deleter
  def atributo(self):
    """
    deleter : se llama al hacer : del obj.atributo
    """
    del self._atributo

  def _validar(self, valor):
    # Logica de validacion
    if valor < 0 :
      raise ValueError()

````

````python
c = Clase(12)
````
