# Manejo de Errores y Excepciones

**Analogia** : Escribir codigo sin manejo de excepsiones es como construir un puente sin considerar el viento, los terremotos o el trafico pesado. Puede parecer solido ... hasta que algo inesperado sucede.

- Errores : Son fallas que impiden que el codigo se ejecute o que se produzcan resultados incorrectos.

- Excepcion : Son eventos detectables en tiempo de ejecutcion que interrumpen el flujo normal del programa ... pero que pueden ser capturados y manejados.

Dos principios que guian el manejo de errores:
- **fail fast**
- **Graceful Degradation**

```python
# Ejemplito1 : Evitar que una operacion aritmetica detenga la ejecucion de mi codigo
def dividir(a,b):
  try:
    resultado = a/b
    return resultado
  except ZeroDivisionError:
    print("Error : NO se puede dividir por cero ")
    return None
```

```python
# Ejemplito3 : Leer un archivo de datos y procesar su contenido numerico
def procesar_archivo(ruta):
  try:
    with open(ruta, "r") as f:
      lineas = f.readlines()

    datos = []
    for i, linea in enumerate(lineas, start =1 ):
      linea = linea.strip()
      if not linea:
        continue # ignoramos lineas en blanco

      # Dividir por espacios (manaje espacios en blanco o tabs)
      campos = linea.split()

      # Validemos que cada linea tiene 6 columnas
      if len(campos) != 6 :
        raise ValueError(f"Linea {i} no tiene 6 columnas")

      # Convertir cada campo (6 en cada fila) es un numero float
      fila = [float(campo) for campo in campos]
      datos.append(fila)
    return datos
  except FileNotFoundError:
    print("Error : El archivo no se puede cargar ")
    return []
  except ValueError:
    print("Error : El contenido del archivo es no numerico")
    return []
```

# El bloque try-except-else

Es una extension del manejo basico de excepciones (try-except) que desempeña
un papel crucial en la separacion semantica entre codigo que puede fallar y codigo que debe ejecutarse solo si no fallo nada

```python
# Ejemplito : COnversion segura de entrada con confirmacion
entrada = input("Ingresa un numero punto flotante : ")

try :
  numero = float(entrada)
  resultado = numero**(0.5)
except ValueError:
  print("Entrada invalida")
except TypeError:
  print("Ingreso invalido : Debe ser mayor o igual a cero ")
else:
  print(f"Conversion exitosa : Resultado : %s" %(resultado))
```

# El bloque try-except-else-finally

Esta forma combina cuatro clausulas:
- try : codigo que podria fallar
- except : manejo de errores especificos
- else: logica que solo debe ejecutarse si no hubo excepciones.
- finally : Codigo de limpieza garantizada, que se ejecuta siempre.

```python
# Ejemplito : Division segura con limpieza de variables
def dividir_con_finally(a,b):
  resultado  = None
  try:
    resultado = a/b
    print("Division Realizada")
  except ZeroDivisionError:
    print("Division por cero")
  else:
    print(f"El resultado de dividir {a} sobre {b} es {resultado}")
  finally:
    print("Bloque finally : limpieza completada")
  return resultado
```

```python
# else + finally
dividir_con_finally(5,3)
```

```python
# except + finally
dividir_con_finally(3,0)
```

# Palabra reservada assert

`assert` es documentacion ejecutable


```python
x = -2
assert x>0, "x debe ser positivo"
```

```python
# Ejemplito1 : Validacion de precondicion en una funcion matematica
def raiz_cuadrada(x):
  """
  calcula la raiz cuadrada de x (x >= 0)
  """
  assert x >= 0 , "Precondicion violada : x debe ser no negativo"
  return x**(0.5)
```

```python
# Uso correcto
print(raiz_cuadrada(9))
```

```python
raiz_cuadrada(-4)
```

```python
# Ejemplito2
def fibonacci(n):
  """
  Genera  los primeros n numeros de la sucesion de Fibonacci
  """
  assert n>=0 and isinstance(n,int) , "n debe ser entero no negativo"

  if n == 0 :
    return []
  elif n == 1:
    return [0]

  seq = [0, 1]
  for i in range(2,n):
    siguiente = seq[-1] + seq[-2]
    seq.append(siguiente)
    # Invariante : Cada nuevo elemento es suma de los dos anteriores
    assert seq[i] == seq[i-1] + seq[i-2], "Invariante de Fibonacci"

  return seq
```

```python
fibonacci(-200)
```
