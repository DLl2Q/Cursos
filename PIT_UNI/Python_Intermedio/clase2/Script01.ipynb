# Definición de excepciones en Python

**raise** : Lanza una excepcion explicitamente

**assert** : Verifica una condicion, y lanza un excepcion automaticamente (AssertionError) si la condicion es falsa

  - Se usa para verificar suposiciones del programador durante pruebas o desarrollo

  - Se desactiva en caso de ejecucion optimizada (Si en una terminal usamos `python -O mi_codigo.py` esa opcion `-O` desactiva los `assert`)

  - No se recomienda para validar entradas de usuario o datos externos.

**try-except** : permite capturar y manejar excepciones

**try-except-else** : Agrega un bloque else que se ejecuta solo si no hay excepciones

**try-except-else-finally** : El bloque finally se ejecuta siempre , ocurra o no una excepcion

# Definicion de Excepciones Personalizadas

```python
# Una buena recomendacion es que mis clases sean clases hijas de Exception
class MiExcepcion(Exception):
  """
  Documentacion mediante un docstring
  Descripcion del error
  """
  pass
```

```python
# Para lanzar o ejecutar nuestra excepcion personalizada
raise MiExcepcion("Mensaje descriptivo")
```

# Ejemplos

```python
# Ejemplito1 : Definamos una excepcion personalizada para validar edad
# Si la edad es negativa (o si es mayor que 115) se lanza una excepcion

class EdadInvalidaError(Exception):
  """
  Se lanza cuando la edad sea negativo o mayor que 115
  """
  pass

def validar_edad(edad):
  if (edad < 0) or (edad >= 115 ):
    raise EdadInvalidaError(f"Edad Invalida. {edad}. Debe encontrarse en [0,115]")
  return f"Edad Valida : {edad}"
```

```python
# Pruebas : Usemos el bloque try-except

try :
  print(validar_edad(25))
  print(validar_edad(-2))
except EdadInvalidaError as e:
  print(f"Error capturado {e}")
```

```python
# Ejemplito2 : Division segura con excepcion personalizada
# Definamos una excepcion llamada DivisionPorCeroError
# Usaremos esta nueva excepcion personalizada en lugar de ZeroDivisionError

class DivisionPorCeroError(Exception):
  """
  Division por cero no permitida
  """
  pass

def dividir_seguro(a,b):
  if b == 0:
    raise DivisionPorCeroError("PIT427 : 'No se puede dividir por cero'")
  return a/b
```

```python
# Algunas pruebas
try :
  print(dividir_seguro(10,2))
  print(dividir_seguro(10,0))
except DivisionPorCeroError as e:
  print(f"Error {e}")
```

```python
# Ejemplito3 : Validacion de email
# Definir una excepcion llamada EmailInvalidoError
# Valida que un email contenga "@" y "."

class EmailInvalidoError(Exception):
  """
  Email no cumple con el formato basico
  """
  pass

def validar_email(email):
  if "@" not in email or "." not in email:
    raise EmailInvalidoError(f"Email invalido '{email}'")
  return "Email valido"
```

```python
# Algunas pruebas
try:
  print(validar_email("usuario@pit427.uni.edu.pe"))
  print(validar_email("algo_arroba_invalido"))
except EmailInvalidoError as e :
  print(f"Error : {e}")
```

```python
# Ejemplito4  : Sistema de calificaciones
# Define una excepcion llamada CalificacionInvalidaError
# Valida que una calificacion este en [0,20]

class CalificacionInvalidaError(Exception):
  """
  Calificacion fuera de rango valido
  """
  pass

def registrar_calificacion(calif):
  if not isinstance(calif, (int,float)):
    raise TypeError("La calificacion debe de ser numerica")
  if calif < 0 or calif> 20:
    raise CalificacionInvalidaError(f"Calificacion Invalida '{calif}'")
  return f"Calificacion registrada '{calif}'"
```

```python
# Prueba
try:
  print(registrar_calificacion(12))
  print(registrar_calificacion(55))
except (CalificacionInvalidaError, TypeError) as e:
  print(f"Error {e}")
```

```python
# Ejemplito5 : Jerarquia de errores de base de datos
# Define una jerarquia de excepciones para operaciones de base de datos
# ErrorDB, ErrorConexion
# ErrorDB, ErrorConsulta

class ErrorDB(Exception):
  """
  PIT427 : Error general de la base de datos en todos los proyectos
  """
  pass

class ErrorConexion(ErrorDB):
  """
  Error al conectar a la base de datos
  """
  pass

class ErrorConsulta(ErrorDB):
  """
  Error al ejecutar la consulta SQL
  """
  pass

def conectar_db():
  # Bajo alguna logica de negocio
  raise ErrorConexion("Fallo al conectar : Servidor no disponible")

def ejecutar_consulta():
  # Bajo una logica de negocio
  raise ErrorConsulta("Consulta invalida : Tabla inexistente")
```

```python
# Pruebas
try:
  conectar_db()
except ErrorDB as e:
  print("Rutina de mantenimiento")
  print(f"Error de base de datos : {e}")

try:
  ejecutar_consulta()
except ErrorDB as e:
  print("Rutina de mantenimiento")
  print(f"Error de consulta : {e}")
```

```python
# Ejemplito6 : Validacion de coordenadas geograficas
# Definamos una excepcion : CoordenadaInvalidaError
# Valida que la latitud este entre -90 y 90, y la longitud entre -180 y 180

class CoordenadaInvalidaError(Exception):
  """
  PIT427 : Coordenada fuera de rango valido
  """
  pass

def validar_coordenada(lat,lon):
  if not (-90 <= lat <= 90):
    # Mediante una logica de negocio
    raise CoordenadaInvalidaError(f"Latitud Invalida : {lat}")
  if not (-180 <= lon <= 180):
    # Mediante una logica de negocio
    raise CoordenadaInvalidaError(f"Longitud Invalida : {lon}")
  return "Coordenadas validas"


```

```python
# Pruebas
try:
  print("Alguna operacion previa ...")
  print(validar_coordenada(40.7128, -74.0060))
  print(validar_coordenada(100,-74))
  print(validar_coordenada(-100,159))
except CoordenadaInvalidaError as e :
  print("Rutina de mantenimiento")
  print(f"Error {e}")
```

```python
# Ejemplito7 : Validacion de una matriz cuadrada
# Definimos una excepcion MatrizNoCuadradaError
# Valida que una matriz sea cuadrada (n x n)

class MatrizNoCuadradaError(Exception):
  """
  La matriz debe ser cuadrada
  """
  pass

def es_cuadrada(matriz):
  n = len(matriz)
  # Si el numero de columnas en todas las filas no es igual a n entonces ejecutamos/lanzamos
  # la excepcion MatrizNoCuadradaError
  if not all(len(fila) == n for fila in matriz ):
    raise MatrizNoCuadradaError("La matriz no es cuadrada")
  return True
```

```python
# Pruebas
matriz1 = [[1,2], [3,4]]
matriz2 = [[1,2,3], [9,8,7,6]]

try:
  es_cuadrada(matriz1)
except MatrizNoCuadradaError as e :
  print(f"Matriz1 : Error {e}")


try:
  es_cuadrada(matriz2)
except MatrizNoCuadradaError as e :
  print(f"Matriz2 : Error {e}")
```

```python
# Ejemplito8 : Sistema de reserbas con excepciones anidadas
#  Definamos una jerarquia de excepciones para un sistema de reservas
# ErrorReserva -> AsientoNoDisponible
# ErrorReserva -> FechaInvalida

class ErrorReserva(Exception):
  """
  Error general de reserva
  """
  pass

class AsientoNoDisponible(ErrorReserva):
  """
  Asiento ya reservado
  """
  pass

class FechaInvalida(ErrorReserva):
  """
  Fecha fuera de rango
  """
  pass


def reservar_asiento(asiento, fecha):
  if asiento in ["A1", "B2"]:
    raise AsientoNoDisponible(f"Asiento {asiento} no se encuentra disponible ")
  if fecha < "03/11/2025":
    raise FechaInvalida(f"Fecha seleccionada no es valida")
  return f"Reserva Confirmada : Asiento {asiento} , fecha {fecha}"
```

```python
# Pruebas
try :
  print(reservar_asiento("C3","09/11/2025"))
  print(reservar_asiento("A1", "11/11/2025"))
except ErrorReserva as e:
  print(f"Error de reserva")
```
