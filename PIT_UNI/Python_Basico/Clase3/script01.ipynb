```python
# Dados los tres lados de un triangulo por el usuario (la funcion input)
# Usar la formula de Heron para calcular su area

# Rutinas de entrada
lado_A = float(input("Ingresa el primer lado del triangulo : "))
lado_B = float(input("Ingresa el segundo lado del triangulo : "))
lado_C = float(input("Ingresa el tercer lado del triangulo : "))

# Realizar algunos calculos (para obtener la respuesta buscada)
# Teorico : Formula de Heron
s = (lado_A + lado_B + lado_C)/2
area = (s*(s-lado_A) * (s - lado_B) * (s-lado_C))**(0.5)

# Mostramos los resultados : Rutinas de salida
print("-----------------------------------------------------------")
print("Calculo del area de un triangulo usando la formula de Heron")
print("-----------------------------------------------------------")
print("A : ", lado_A)
print("B : ", lado_B)
print("C : ", lado_C)
print("---------------")
print("Area del triangulo ", area)
```

```python
# %f : punto flotante
# %.2f : punto flotante con dos cifras en la parte decimal
# %d : numero entero  (int)
# %s : cadena de caracteres (str)
# %e : notacion cientifica de un punto flotante
# %.2e : notacion cientifica de un punto flotante con dos cifras significativas
```

# Estructuras de decision

```python
# Escenario1 : Evaluar una condicional y en funcion del resultado
# ejecutar una determinada rutina

cadena_str = input("INgresa una cadena de caracteres : ")

if len(cadena_str) >= 6 :
  print("Ingreso Correcto")
```

```python
# Escenario2 : Ejecutar una sentencia en caso la condicional tenga
# valor booleano de True y otra sentencia en caso la condicional
# tenga valor booleano de False

cadena_str = input("Ingresa una cadena de caracteres : ")

if len(cadena_str) >= 6 :
  print("Ingreso Correcto")
else:
  print("Considera agregar mas caracteres !!!")
```

```python
# Escenario3 : El caso cuando tenemos que evaluar mas de una condicional
# Si edad >1 y menor a 18 => "Es momento de disfrutar"
# Si edad >=18 y menor a 30 => "Es momento de cocechar"
# Si edad >=30 y menor a 50 => "es momento de Construir"
# Si edad >= 50 y menor a 115 => "a descanzar"

edad = int(input("Ingresa  tu edad como un numero entero : "))

if (edad > 1) and (edad <18) :
  print("Es momento de disfrutar")
elif (edad >= 18) and (edad <30) :
  print("Es momento de cocechar")
elif (edad >= 30) and (edad <50):
  print("es momento de Construir")
elif (edad >=50)  and (edad < 115):
  print("a descanzar")
else:
  print("Caso por defecto")
```

```python
# Condiciones para aceptar un password
# Si tiene menos de 8 caracteres => no sirve (no cumple con la longitud minima)
# en caso contrario :
  # Si solo tiene letras o solo tiene numeros => no sirve
  # Si solo tiene mayusculas o solo tiene minusculas => no sirve
  # Si esta compuesto por letras (mayusculas y minusculas) y numeros => Casi sirve
  # Si esta compuesto por letras (mayusculas y minusculas) y numeros  y caracteres especiales => FUERTE

# Rutina de entrada
contraseña = input("Ingresa tu contraseña : ")

# Implementemos la logica requerida
if len(contraseña) < 8 :
  print("No sirve (no cumple con la longitud minima)")
else :
  print("Logica implementada")
  if contraseña.isalnum():
    print("No sirve (Solo contiene letras y/o numeros)")
    if contraseña.isupper() or contraseña.islower() :
      print("No sirve (Solo tiene letras (mayusculas o minusculas) y numeros)")
    else :
      print("Casi Sirce (tiene mayusculas, minusculas y numeros) ")
  else:
      print("FUERTE")
```
