# Desarrollo y evolucion de los Lenguajes de Programacion

A nivel fundamental, un lenguaje de programacion es un sistema formal de comunicacion que permite a un ser humano (programador) especificar una secuencia de instrucciones para que una maquina (computadora) las ejecute.

* **Perspectiva matematica** : Es un lenguaje formal con una sintaxis (reglas de formacion de sentencia) y una semantica (significado de esas sentencias) bien definidas. Busca ser no ambiguo, a diferencia del lenguaje natural.

$$
L  = (\Sigma , R, S)
$$

Donde:  
* $\Sigma$ : es el alfabeto (simbolos permitidos)
* $R$ : es el conjunto de reglas de produccion (gramtica)
* $S$ : Los simbolos

* **Perspectiva de Ingenieria de Software** : Es una herramienta de Abrstraccion. Nos permite expresar soluciones a problemas complejos sin tener que preocuparnos por los detalles de bajo nivel del hardware.

# Niveles de Abstraccion : La Torre de Babel computacional

* Lenguaje de Alto Nivel (Python, Java, C#)
  * Sintaxis cercana al lenguaje humano
  * Alta portabilidad entre sistemas
  * Gestionan la memoria automaticamente (Garbage Collection)
  * Foco: Productividad del desarrollador

* Lenguajes de Nivel medio (C, C++)
  - Combinan abstracciones de alto nivel con acceso directo a la memoria
  - Ofrecen un control granular
  - Base para sistemas operativos, drivers y motores de videojuegos
  - Foco : Rendimiento y Control

* Lenguajes de Bajo Nivel (Ensamblador)
  - Representacion mnemonica del codigo maquina
  - Especifico para cada arquitectura de CPU (x86, ARM)
  - Usado para optimizaciones extremas y programacion de sistemas embebidos
  - Foco : Interaccion directa con el hardware.
