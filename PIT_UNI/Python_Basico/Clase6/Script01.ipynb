# El dato de tipo `dict` (diccionario)

```python
# Definicion : Se definen con llaves , los elementos se  separan por comas
# Pero, cada elemento esta compuesto por un par llave:valor

import math as m
import random as rnd

d1 = {}
d2 = dict()
d3 = {"A": (1,2,3,"#",5)}
d4 = {"one":rnd.randint(1,6), "two":rnd.random(), "three":list(range(6,25,3))}
```

```python
# Observacion
# Las llaves o claves de un dict deben ser objetos hashables : str, int, float, tuple
# Los tipos de datos mutables no pueden ser claves de un dict en python
# Los valores de un dict pueden ser de cualquier tipo : funciones, clases, otro dicts,etc
```

```python
# Accedamos all valor que corresponde a la llave "three" del objeto (tipo dict) d4
d4["three"]
```

```python
# Union (|) :Disponible desde la version 3.9+
# Permite fusionar diccionarios
# Si hay claves duplicadas, los valores del operando derecho prevalecen

d5 = {"a":1, "b":2}
d6 = {"b":3.1415, "c":"666"}
d7 = d5 | d6
d7
```

```python
# Asignacion con union (|=) : disponible desde la version 3.9+
# Actualiza el diccionario "IN-PLACE" con otro
d8 = {"a": 123, "b":987}
d9 ={"b":789, "c":0}
d8 |= d9
d8
```

```python
# Operaciones de pertenencia
# in : devuelve True si la clave/llave esta presente en el diccionario
# not in : Devuelve True si la clave/llave no esta presente en el diccionario

d10 = {"nombre":"Carlos",
       "edad":30,
       "Profesion": "Matematico"}

print("nombre" in d10)
print("direccion" not in d10)
```

```python
# Comparacion de igualdad (== , !=)

d11 = {"peso":101, "estatura" : 1.85}
d12 = {"estatura":1.85, "peso":101}
d11 == d12

# Regla : Dos diccionarios son iguales si contienen exactamente los mismos pares clave:valor
# independiente del orden de insercion
```

## El metodo setdefault

```python
# Documentacion
help({}.setdefault)
```

```python
# COntemos las ocurrencias de valores en una lista y almacenaremos esta informacion
# en un diccionario

# Definamos una lista de palabras
palabras = ["manzana", "pera", "manzana", "uva", "pera", "manzana"]
conteo = {}

for palabra in palabras:
  conteo.setdefault(palabra, 0)
  conteo[palabra] = conteo[palabra] +1

print(conteo)

```

```python
# Agrupar elementos
estudiantes = [
    ("Ana", "Matematicas"),
    ("Luis", "Fisica"),
    ("Carla", "Matematicas"),
    ("Diego", "Fisica")
    ]

grupos = {}
for nombre,materia  in estudiantes:
  grupos.setdefault(materia, []).append(nombre)

grupos
```

```python
# Inicializar configuracion con valores por defecto
config = {"modo" : "produccion"}

# Asegurar que ciertas claves/llaves existan
puerto = config.setdefault("puerto",8000)
debug = config.setdefault("debug",False)

print(config)
```

```python
# Evitar errores al acceder a claves/llaves anidadas (nivel1)
usuario = {"nombre" : "Vanessa"}

# Asegurar que exista la clave "perfil"
perfil = usuario.setdefault("perfil", {})
perfil["edad"] = 30

print(usuario)
```

```python
# Combinar con entrada de usuario

carrito = {}

producto = input("Producto: ")
carrito.setdefault(producto, 0)
carrito[producto] = carrito[producto] +1

print(carrito)
```

```python
# Cuanto usar setdefault :
  # Cuanto necesitas inicializar una clave solo si no existe
  # En bucles donde vas a modificar el valor asociado
  # Como alternativa ligera a collections.defaultdict
```

## El metodo update

```python
# Documentacion
help({}.update)
```

```python
# Actualizar con otro diccionario
usuarios = {"nombre":"john", "edad":24}
nuevos_datos = {"edad":29, "ciudad":"Pacasmayo"}

usuarios.update(nuevos_datos)
print(usuarios)
```

```python
# Actualizamos el diccionario con una lista de tuplas
d16 = {"a":1}
d16.update([("b", 12),("c",99),("d","0")])
d16
```

```python
# Usemos argumentos con nombre (keyword arguments)
d17 = {"Eva1":3}
d17.update(Eval2= 5, Eval3=8)
d17
```
