```python
# Otro Ejemplito : Sumar los numeros :  1,2,3,...,20

# suma : variable donde se ira acumulando la suma de elementos
suma = 0

# Utilizando una estructura for para acumular los elementos del objeto range en el acumulador : suma
for i in range(1,21):
  suma = suma + i

# Podemos mostrar la suma de numeros de 1 a 20
print("La suma : 1+2+3+...+20 = ", suma )
```

```python
# Otro ejemplito : Calculo de la productoria de los numeros : 4,5,6,7,8,9,...,15

# prod = variable donde se ira acumulando el producto de elementos
prod = 1

# Utilizando una estructura repetitiva for vamos a acumular los elemento de un objeto range : range(4,16)
for j in range(4,16):
  prod = prod * j

# Mostremos los resultados
print("Productoria(4*5*6*...*15) : %d" %(prod))
```

```python
# Otro Ejemplito : Primalidad

# Solicitemos al usuario un numero entero mayor que 1
n = int(input("Ingresa un numero entero mayor que 1  : "))

# Por ejemplo :
# n = 33
# vamos a calcular el resto de dividir n por cada numero entre 2 y 32
# i = 2 : 33 % 2 != 0
# i = 3 : 33 % 3 == 0 -> i = 3 es un divisor de n = 33 -> n = 33 no es primo (sentencia break : salir del bucle )
# i = 4 : 33 % 4 != 0
# i = 5 : 33 % 5 != 0

# Veamos la implementacion

# Iniciar suponiendo que el valor n (ingresado por el usuario es primo)
es_primo = True

# Un numero menor o igual a 1 no es primo
if n <= 1 :
  es_primo = False
else:
  # Iterar desde 2 hasta n-1 para buscar divisores
  for i in range(2,n):
    # El si numero n (ingresado por el usuario) es divisible por algun numero de el objeto range => no es primo
    if (n % i == 0):
      es_primo = False
      # Detenemos la estructura repetitiva for : sentencia break
      break

# Mostremos el resultado
if es_primo:
  print(n , " es un numero primo ")
else:
  print(n, " no es un numero primo ")
```

# Estructura de repeticion `WHILE`

```python
# Ejemplito : Pedir al usuario que ingrese un numero positivo, si ingresa un
# numero <= 0 le volvemos a pedir que ingrese el numero
# ACTUALIZACION DEL REQUERIMIENTO : Necesito contar el numero de veces que el
# usuario ingreso un numero <= 0
num_veces_menor_igual_cero = 0

numero = int(input("Ingresa un numero positivo : "))

while numero<=0:
  num_veces_menor_igual_cero = num_veces_menor_igual_cero +1
  numero = int(input("Ingresa un numero positivo"))

# Puedo mostrar al usuario el numero de ingresos incorrectos
print("Numero de ingresos incorrectos : ", num_veces_menor_igual_cero)
```

```python
# Otro Ejemplito : Implementemos la suma de numeros de 1 a 20 usando
# una estructura repetitiva while

suma = 0
num=1
while num <=20:
  suma = suma + num
  # print("Suma Parcial  : ", suma)
  num = num +1

# Mostrar el resultado
print("La suma total de numero de 1...20 es : ", suma)
```

```python
# Otro ejemplito : Contar cuantos numeros pares ingresa el usuario (hasta que ingrese un cero)

# NUmero de pares ingresados por el usuario
pares = 0

# Iniciamos con un numero impar
n = 3


while n != 0:
  n = int(input("Ingresa un numero (0 para salir) : "))
  if ( n%2 == 0 ) and (n!=0):
    pares = pares +1

print("Numero total de pares ingresados : ", pares)
```

```python
# Otro ejemplito : Simulemos el lanzamiento de un dado hasta que salga 5
# ACTUALIZACION DE REQUERIMIENTO : Contar el numero de veces que sale un numero
# diferente de 5
num_veces_diferente_cinco = 0

result_exp = 0
while result_exp != 5:
  num_veces_diferente_cinco = num_veces_diferente_cinco +1
  result_exp = rnd.randint(1,6)
  print("Salio : ", result_exp)

# Si se detiene la estructura repetitiva while => salio 5 !!!
print("Ganamos")
print("Numero de lanzamientos : ", num_veces_diferente_cinco-1)
```

```python
# Otro ejemplito : Calcular el promedio de los numeros ingresados por el usuario
# hasta que el usuario diga "FIN"

suma = 0
conteo = 0
entrada = ""

while entrada != "FIN":
  entrada = input("Ingresa un numero o 'FIN' para terminar : ")
  if entrada != "FIN" :
    suma = suma + float(entrada)
    conteo = conteo + 1

# Mostrar el resultado
if conteo>0 :
  print("Promedio : ", suma/conteo)
```

```python
# Otro ejemplito : Verificar si un numero es primo (version WHILE)

n = int(input("Ingresa un numero entero positivo : "))
divisor = 2
es_primo = True

while divisor <n :
  if n%divisor == 0:
    es_primo = False
    break
  divisor = divisor +1

if es_primo and (n>1):
  print(n , " es primo ")
else :
  print(n, "no es primo ")
```
